```python
import matplotlib.pyplot as plt

# Dados de exemplo (categoria e quantidade)
emocoes = ["Tristeza", "Alegria", "Medo", "Nojo", "Raiva"]
quantidade = [65.7, 31.08, 56.87, 45.65, 38.75]

# Plotar o gráfico de barra
plt.bar(emocoes, quantidade)

# Personalização do gráfico
plt.title('Analise sentimentos sobre noticia de morte de homem Gay em SP')
plt.xlabel('emocoes')
plt.ylabel('Quantidade')


# Mostrar o gráfico
plt.show()
```
